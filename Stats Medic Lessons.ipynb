```python
unit_list = ['https://www.statsmedic.com/ced-apstats-unit1',
            'https://www.statsmedic.com/ced-apstats-unit2',
            'https://www.statsmedic.com/ced-apstats-unit3',
            'https://www.statsmedic.com/ced-apstats-unit4',
            'https://www.statsmedic.com/ced-apstats-unit5',
            'https://www.statsmedic.com/ced-apstats-unit6',
            'https://www.statsmedic.com/ced-apstats-unit7',
            'https://www.statsmedic.com/ced-apstats-unit8',
            'https://www.statsmedic.com/ced-apstats-unit9']
```

```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
link_list = []
for unit in unit_list:
    page = requests.get(unit)
    soup = BeautifulSoup(page.content, 'lxml')
    for url in soup.find_all('div', class_='txtNew', id=re.compile(r'comp-*'))[4].find_all('a'):
        link_list.append(url['href'])
```

```python
link_list
```

```python
pdf_links = []
for lesson in link_list:
    page = requests.get(lesson)
    soup = BeautifulSoup(page.content, 'lxml')
    urls = soup.find_all('a', href=re.compile(r'https://stats-medic1.memberspace.com/content/*'))
    if len(urls) > 2:
        print(lesson)
        pdf_links.append({'name': lesson.replace('https://www.statsmedic.com/ced-apstats-',''), 'url': urls[1]['href']})
```

```python
pdf_links
```

```python
import os
folder_name = 'statsmedic_lessons'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
```

```python
username =  'username',
password = 'password'

with requests.Session() as s:
    site = s.get("https://stats-medic1.memberspace.com/member/sign_in")
    bs_content = BeautifulSoup(site.content, "html.parser")
    token = bs_content.find("input", {"name":"authenticity_token"})["value"]
    login_data = {"email":username,"password":password, "csrf_token":token}
    s.post("https://stats-medic1.memberspace.com/member/sign_in",login_data)
    for item in pdf_links:
        print(item['name'])
        req2 = s.get(item['url'], allow_redirects=True)
        with open(folder_name + '/' + item['name'] + '.pdf', 'wb') as f:
            f.write(req2.content)

```
